```python
import numpy as np 
linearData = np.array([1,2,3,4,5])
quadData = linearData**2
plt.figure()
plt.plot(linearData, '--r')
```

```python
plt.plot(quadData,"-o")
```

```python
plt.gca().fill_between(range(len(linearData)),
                      linearData, quadData,
                      facecolor="#6200EE",
                      alpha=0.005)
```

```python
import pandas as pd
plt.figure()
obvDates = np.arange('2018-12-01', '2018-12-06', dtype='datetime64[D]' )
ovbDates = list(map(pd.to_datetime, obvDates))
plt.plot(obvDates, linearData, '-o', obvDates, quadData, '-o')
```

```python
for item in plt.gca().xaxis.get_ticklabels():
    item.set_rotation(45)
```
